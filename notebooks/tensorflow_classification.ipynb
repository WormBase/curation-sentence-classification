```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D, Flatten, Dropout
from tensorflow.keras.layers import TextVectorization
import numpy as np
```

```python
otherexpr_data = [line.rstrip("\n").split("\t") for line in open("../gold_standard_sentences/clean_with_negative_otherexpr.csv")]
otherexpr_data = [row for row in otherexpr_data if int(row[-1]) < 8 or "express" in row[2].lower()]
kinaseact_data = [line.rstrip("\n").split("\t") for line in open("../gold_standard_sentences/clean_with_negative_kinaseact.csv")]
kinaseact_data = [row for row in kinaseact_data if int(row[-1]) < 8 or "kinase" in row[2].lower()]
```

```python
from sentence_transformers import SentenceTransformer
embedder = SentenceTransformer('all-MiniLM-L6-v2')
corpus_expr = embedder.encode([row[2] for row in otherexpr_data], convert_to_tensor=True).numpy()
corpus_kinaseact = embedder.encode([row[2] for row in kinaseact_data], convert_to_tensor=True).numpy()
```

```python
labels_expr = np.asarray([1 if int(row[-1]) < 8 else 0 for row in otherexpr_data])
labels_kinaseact = np.asarray([1 if int(row[-1]) < 8 else 0 for row in kinaseact_data])
```

```python
from sklearn.model_selection import train_test_split

data_train, data_test, labels_train, labels_test = train_test_split(corpus_expr, labels_expr, test_size=0.20)
data_val, data_test, labels_val, labels_test = train_test_split(data_test, labels_test, test_size=0.50)
```

```python
model_expr = Sequential()
#model.add(Flatten())
#model.add(Dropout(0.6))
#model.add(Dense(10, activation='relu'))
model_expr.add(Dense(1, activation='sigmoid'))
# compile the model
model_expr.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', 'Precision', 'Recall'])
# summarize the model
# fit the model
history_expr = model_expr.fit(data_train, labels_train, validation_data=(data_val, labels_val), epochs=65)
```

```python
print("Expression classifier")
print("Test Set Size: ", len(data_test))
loss, accuracy, precision, recall = model_expr.evaluate(data_test, labels_test)
print('Accuracy: %f' % (accuracy*100))
```

```python
model_kinase = Sequential()
#model.add(Flatten())
#model.add(Dropout(0.6))
#model.add(Dense(10, activation='relu'))
model_kinase.add(Dense(1, activation='sigmoid'))
# compile the model
model_kinase.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', 'Precision', 'Recall'])
data_train, data_test, labels_train, labels_test = train_test_split(corpus_kinaseact, labels_kinaseact, test_size=0.20)
data_val, data_test, labels_val, labels_test = train_test_split(data_test, labels_test, test_size=0.50)
history = model_kinase.fit(data_train, labels_train, validation_data=(data_val, labels_val), epochs=65)
```

```python
print("Kinase activity classifier")
print("Test Set Size: ", len(data_test))
loss, accuracy, precision, recall = model_kinase.evaluate(data_test, labels_test)
print('Accuracy: %f' % (accuracy*100))
```
